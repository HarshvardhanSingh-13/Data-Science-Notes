```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
df.head()
```

```python
df.info()
```

## Data Cleaning

```python
## missing values
df[df.isnull().any(axis=1)]
```

The dataset is converted into 2 sets based on Region from 122th Index, we can make a new column based on the Region 
1) "Bejiaia Region Dataset"
2) "Sidhi-Bel Abbes Region Dataset"

```python
df.loc[:122,"Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df
```

```python
df.isnull().sum()
```

```python
## Remove the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix Spaces in Column name
df.columns = df.columns.str.strip()
```

```python
df.columns
```

### Change the required column as Integer Data Types

```python
df.head()
```

```python
df.columns
```

```python
df[['month','day','year','Temperature','RH',"Ws"]] = df[['month','day','year','Temperature','RH',"Ws"]].astype(int)
```

```python
df.info()
```

```python
### Changing the other column to float Data Type
objects = [i for i in df.columns if df[i].dtypes == 'O']
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## Lets Save the Cleaned Dataset
df.to_csv('Algerian_dataset_Cleaned.csv',index=False)
```
